```python
import os
from PIL import Image
import wave

# Path to the folder containing the original images
input_folder = 'C:\\Users\\ppk_2\\Desktop\\College\\DeepLearning_Lab\\Steganogrpahy\\Audio\\Train'

# Path to the folder where you want to save the embedded images
output_folder = 'C:\\Users\\ppk_2\\Desktop\\College\\DeepLearning_Lab\\Steganogrpahy\\Audio\\Embedded'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# List all the image files in the input folder
image_files = [f for f in os.listdir(input_folder) if f.endswith('.jpg')]

for i, image_file in enumerate(image_files):
    # Load the original image
    original_image = Image.open(os.path.join(input_folder, image_file))

    # Load the audio file (you can replace 'audio.wav' with your audio file)
    audio = wave.open('C:\\Users\\ppk_2\\Desktop\\College\\DeepLearning_Lab\\Steganogrpahy\\Audio\\test.wav', 'rb')
    audio_data = audio.readframes(audio.getnframes())

    # Calculate image capacity
    width, height = original_image.size
    num_pixels = width * height
    bits_per_pixel = 3  # 3 bits per pixel for RGB
    total_bits_capacity = num_pixels * bits_per_pixel
    audio_capacity_bytes = total_bits_capacity // 8

    # Check if audio data can be embedded in the image
    if len(audio_data) <= audio_capacity_bytes:
        print(f"Embedding audio into {image_file}...")

        # Embed audio data into the image
        embedded_image_data = []
        audio_index = 0
        image_data = list(original_image.convert('RGB').getdata())

        for r, g, b in image_data:
            # Make sure audio_index stays within the bounds of audio_data
            audio_index %= len(audio_data)

            # Embed audio data into the least significant bits of each color channel
            r_embed = r & 0xFE | (audio_data[audio_index] & 1)
            g_embed = g & 0xFE | ((audio_data[audio_index] >> 1) & 1)
            b_embed = b & 0xFE | ((audio_data[audio_index] >> 2) & 1)

            embedded_image_data.append((r_embed, g_embed, b_embed))
            audio_index += 1

        # Create a new image with the embedded data
        embedded_image = Image.new('RGB', original_image.size)
        embedded_image.putdata(embedded_image_data)

        # Save the embedded image with the name "emb[i].jpg"
        embedded_image.save(os.path.join(output_folder, f'emb{i}.jpg'))

    else:
        print(f"Skipping {image_file} as it is not compatible for embedding.")
        # You can add code here to delete the incompatible image if needed.
```
